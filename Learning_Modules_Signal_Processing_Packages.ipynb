Example: itertools.filterfalse(predicate, iterable)

Predicate: any function that returns a true or false

Example:  Filter out numbers that are not prime from the list

```python
def is_prime(n):
    for i in range(2,n):
        if n % i == 0:
            return False
    return True
```

```python
is_prime(7)
```

```python
is_prime(20)
```

```python
it.filterfalse(is_prime, range(20))
```

```python
list(it.filterfalse(is_prime, range(20)))
```

```python
list(filter(is_prime, range(20)))
```

Collection Module

Module for implementing specilaized container types

Example: collections.deque([iterable[,maxLen]])


List is very efficient for adding items to the end of list but adding at the begiining is very inefficient. Prepending is very inefficient. 


deque is very useful in such cases

```python
from collections import deque
```

```python
dir(deque)
```

```python
myiter = range(10, 40, 2)
buffer = deque(myiter, 8)
buffer
```

```python
buffer.append(99)
```

```python
buffer
```

```python
buffer.appendleft(100)
buffer
```

```python
z = buffer[5]
```

```python
z
```

```python
type(z)
```

Different from Lists, we cannot do slicing or range operation on deque

```python
z1 = buffer[4:6]
```
